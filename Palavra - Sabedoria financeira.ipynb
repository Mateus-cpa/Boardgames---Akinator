Um lenhador não recebe seu salário apenas por afiar seu machado. Apesar de este ser necessário para cortar a árvore e que um bom fio facilitará o trabalho duro, mas somente o planejamento não assegura a execução.

>⁴ Quem observa o vento, nunca semeará, e o que olha para as nuvens nunca segará.

# 4. Investimentos

## O primeiro investimento
1 Reis 3:5-13 (NVI)
O crescimento profissional e da renda gerada por ela é fruto de boas escolhas. Salomão escolheu a sabedoria antes de todas as riquezas.

## Reserva de emergência
José do Egito nos ensina sobre a regra dos 20%. Ele inspirou as dicas atuais da divisão 50-30-20, onde dividimos as despesas em 50% para necessidades, 30% para despesas não essenciais e 20% para reservas e investimentos.

Nunca se sabe o dia de amanhã, pois quanso o amanhã chega, precisamos estar preparados.

## Diversificação
Eclesiastes 11:2 (NVI):
> Reparta o que você tem com sete, até mesmo com oito, pois você não sabe que desgraça poderá cair sobre a terra.

Seus investimentos de longo prazo devem ser divididos em diferentes formas de ganhar dinheiro:
- Empresas (seja um empresa familiar, ações, fundos de investimentos)
- Imóveis (seja em imóveis físicos, cotas de fundos imobiliários, letras de crédito - LCA/LCI, certificados de recebíveis - CRA/CRI)
- Reservas de valor (dinheiro físico, outras moedas correntes, títulos governamentais, Certificados de Depósito Bancário - CDB)

Se colocar tudo em um único investimento aumentam os riscos. Pode gerar bons lucros, mas podem causar grandes prejuízos. Diversificar é se proteger.



# 5. Propósito

1 Timóteo 6:10 (NVI)
>⁹ Os que querem ficar ricos caem em tentação, em armadilhas e em muitos desejos descontrolados e nocivos, que levam os homens a mergulharem na ruína e na destruição,
¹⁰ pois o amor ao dinheiro é raiz de todos os males. Algumas pessoas, por cobiçarem o dinheiro, desviaram-se da fé e se atormentaram a si mesmas com muitos sofrimentos.

Enriquecimento é fruto do amadurecimento dos passos anteriores: Planejamento, Receita, Despesa e Investimento.

Dinheiro é uma ferramenta, não a prioridade.

A pessoa que prioriza o dinheiro ela troca sua vida (tempo, saúde, relacionamento, paz e sono), e por que não dizer sua alma, por ele.

## Caridade

O equilíbrio entre caridade e crescimento deixa um legado duradouro. Riqueza com propósito garante provisão não só para sua família, mas para outras pessoas.
